```python
#Time Conversion

#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'timeConversion' function below.
#
# The function is expected to return a STRING.
# The function accepts STRING s as parameter.
#

def timeConversion(s):
    # Write your code here
    
    
    if s[:2] == "12":
        if s[-2:] == "AM":
            hh = "00"
            s = hh + s[2:8]
        else:
            hh = "12"
            s = hh + s[2:8]
    else:
        if s[-2:] == "PM":
            hh = str(int(s[:2]) + 12)
            s = hh + s[2:8]
        else:
            s = s[:8]
    return s
    

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    s = input()

    result = timeConversion(s)

    fptr.write(result + '\n')

    fptr.close()
```

```python
#Simple Array Sum

#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'simpleArraySum' function below.
#
# The function is expected to return an INTEGER.
# The function accepts INTEGER_ARRAY ar as parameter.
#

def simpleArraySum(ar):
    # Write your code here
    result = 0
    for i in ar:
        result += i
    return result
        

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    ar_count = int(input().strip())

    ar = list(map(int, input().rstrip().split()))

    result = simpleArraySum(ar)

    fptr.write(str(result) + '\n')

    fptr.close()
```

```python
#Compare the Triplets

#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'compareTriplets' function below.
#
# The function is expected to return an INTEGER_ARRAY.
# The function accepts following parameters:
#  1. INTEGER_ARRAY a
#  2. INTEGER_ARRAY b
#

def compareTriplets(a, b):
    # Write your code here
    alice = 0
    bob = 0
    
    for i in range(3):
        if a[i] < b[i]:
            bob += 1
        elif a[i] > b[i]:
            alice += 1
    
    return [alice, bob]

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    a = list(map(int, input().rstrip().split()))

    b = list(map(int, input().rstrip().split()))

    result = compareTriplets(a, b)

    fptr.write(' '.join(map(str, result)))
    fptr.write('\n')

    fptr.close()
```

```python
# Diagonal Difference

#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'diagonalDifference' function below.
#
# The function is expected to return an INTEGER.
# The function accepts 2D_INTEGER_ARRAY arr as parameter.
#

def diagonalDifference(arr):
    # Write your code here
    sum1 = 0
    sum2 = 0
    
    for i in range(len(arr)):
        
        x = y = i
        sum1 += arr[x][y]
        
    for i in range(len(arr)):
        x = i
        y = len(arr)-1 -i
        sum2 += arr[x][y]

    return abs(sum1 - sum2)
        
        
    

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input().strip())

    arr = []

    for _ in range(n):
        arr.append(list(map(int, input().rstrip().split())))

    result = diagonalDifference(arr)

    fptr.write(str(result) + '\n')

    fptr.close()
```

```python
# Plus Minus

#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'plusMinus' function below.
#
# The function accepts INTEGER_ARRAY arr as parameter.
#

def plusMinus(arr):
    # Write your code here
    tot = len(arr)
    ratios = [0,0,0]
    
    for i in arr:
        if i > 0:
            ratios[0] += 1
        elif i < 0:
            ratios[1] += 1
        else:
            ratios[2] += 1
            
    for i in ratios:
        print("{:.6f}".format(i/tot))

if __name__ == '__main__':
    n = int(input().strip())

    arr = list(map(int, input().rstrip().split()))

    plusMinus(arr)
```

```python
# Staircase

#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'staircase' function below.
#
# The function accepts INTEGER n as parameter.
#

def staircase(n):
    # Write your code here
    space = n-1
    hsh = 1
    for i in range(n):
        print(space*" " + hsh*"#")
        space -= 1
        hsh += 1

if __name__ == '__main__':
    n = int(input().strip())

    staircase(n)
```

```python
#Mini-Max Sum

#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'miniMaxSum' function below.
#
# The function accepts INTEGER_ARRAY arr as parameter.
#

def miniMaxSum(arr):
    # Write your code here

    arrsum = sum(arr)
    
    max = arr[0]
    min = arr[0]
    
    for i in range(len(arr)):
        if arr[i] > max:
            max = arr[i]
        elif arr[i] < min:
            min = arr[i]
    
    minSum = arrsum - max
    maxSum = arrsum - min
    
    print(minSum, maxSum)

if __name__ == '__main__':

    arr = list(map(int, input().rstrip().split()))

    miniMaxSum(arr)
```

```python
# Birthday Cake Candles

#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'birthdayCakeCandles' function below.
#
# The function is expected to return an INTEGER.
# The function accepts INTEGER_ARRAY candles as parameter.
#

def birthdayCakeCandles(candles):
    # Write your code here
    cmax = 0
    for i in candles:
        if i > cmax:
            cmax = i
    count = 0
    for i in candles:
        if i == cmax:
            count += 1
    
    return count
        

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    candles_count = int(input().strip())

    candles = list(map(int, input().rstrip().split()))

    result = birthdayCakeCandles(candles)

    fptr.write(str(result) + '\n')

    fptr.close()
```

```python
#CamelCase

#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'camelcase' function below.
#
# The function is expected to return an INTEGER.
# The function accepts STRING s as parameter.
#

def camelcase(s):
    # Write your code here
    counter = 1
    for c in s:
        if c.isupper() == True:
            counter += 1
    return counter

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    s = input()

    result = camelcase(s)

    fptr.write(str(result) + '\n')

    fptr.close()
```

```python
# Strong Password

#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'minimumNumber' function below.
#
# The function is expected to return an INTEGER.
# The function accepts following parameters:
#  1. INTEGER n
#  2. STRING password
#

def minimumNumber(n, password):
    # Return the minimum number of characters to make the password strong
    a = b = p = d = 1
    for c in password:

        if c in "0123456789":
            a = 0
        if c in "abcdefghijklmnopqrstuvwxyz":
            b = 0
        if c in "ABCDEFGHIJKLMNOPQRSTUVWXYZ":
            p = 0
        if c in "!@#$%^&*()-+":
            d = 0
        
    counter = a+b+p+d
    
    while (n + counter) < 6:
        counter += 1
    
        
    return counter
if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input().strip())

    password = input()

    answer = minimumNumber(n, password)

    fptr.write(str(answer) + '\n')

    fptr.close()
```

```python
#Alternate Characters

#!/bin/python3

import math
import os
import random
import re
import sys
import itertools

#
# Complete the 'alternate' function below.
#
# The function is expected to return an INTEGER.
# The function accepts STRING s as parameter.
#

def alternate(s):
    # Write your code here
    uniq = []
    
    for i in s:
        if i not in uniq:
            uniq.append(i)
    
    if len(uniq) == 1:
        return 0
    
    else:
        combinations = list(itertools.combinations(uniq, len(uniq)-2))
        
        validTs = []
                
        for combo in combinations:
            
            t = s
            
            for char in combo:
                t = t.replace(char, "")
                
            flag = "up"    
            for i in range((len(t)-1)):
                if t[i] == t[i+1]:
                    flag = "down"
                    break
                
            if flag == "up":
                validTs.append(t)
        
        if validTs == []:
            return 0
        else:
            return max(map(len, validTs))
                
    

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    l = int(input().strip())

    s = input()

    result = alternate(s)

    fptr.write(str(result) + '\n')

    fptr.close()
```

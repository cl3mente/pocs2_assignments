```python
#Time Conversion

#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'timeConversion' function below.
#
# The function is expected to return a STRING.
# The function accepts STRING s as parameter.
#

def timeConversion(s):
    # Write your code here
    
    
    if s[:2] == "12":
        if s[-2:] == "AM":
            hh = "00"
            s = hh + s[2:8]
        else:
            hh = "12"
            s = hh + s[2:8]
    else:
        if s[-2:] == "PM":
            hh = str(int(s[:2]) + 12)
            s = hh + s[2:8]
        else:
            s = s[:8]
    return s
    

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    s = input()

    result = timeConversion(s)

    fptr.write(result + '\n')

    fptr.close()
```

```python
#Compare the Triplets

#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'compareTriplets' function below.
#
# The function is expected to return an INTEGER_ARRAY.
# The function accepts following parameters:
#  1. INTEGER_ARRAY a
#  2. INTEGER_ARRAY b
#

def compareTriplets(a, b):
    # Write your code here
    alice = 0
    bob = 0
    
    for i in range(3):
        if a[i] < b[i]:
            bob += 1
        elif a[i] > b[i]:
            alice += 1
    
    return [alice, bob]

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    a = list(map(int, input().rstrip().split()))

    b = list(map(int, input().rstrip().split()))

    result = compareTriplets(a, b)

    fptr.write(' '.join(map(str, result)))
    fptr.write('\n')

    fptr.close()
```

```python
# Diagonal Difference

#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'diagonalDifference' function below.
#
# The function is expected to return an INTEGER.
# The function accepts 2D_INTEGER_ARRAY arr as parameter.
#

def diagonalDifference(arr):
    # Write your code here
    sum1 = 0
    sum2 = 0
    
    for i in range(len(arr)):
        
        x = y = i
        sum1 += arr[x][y]
        
    for i in range(len(arr)):
        x = i
        y = len(arr)-1 -i
        sum2 += arr[x][y]

    return abs(sum1 - sum2)
        
        
    

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input().strip())

    arr = []

    for _ in range(n):
        arr.append(list(map(int, input().rstrip().split())))

    result = diagonalDifference(arr)

    fptr.write(str(result) + '\n')

    fptr.close()
```

```python
#Mini-Max Sum

#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'miniMaxSum' function below.
#
# The function accepts INTEGER_ARRAY arr as parameter.
#

def miniMaxSum(arr):
    # Write your code here

    arrsum = sum(arr)
    
    max = arr[0]
    min = arr[0]
    
    for i in range(len(arr)):
        if arr[i] > max:
            max = arr[i]
        elif arr[i] < min:
            min = arr[i]
    
    minSum = arrsum - max
    maxSum = arrsum - min
    
    print(minSum, maxSum)

if __name__ == '__main__':

    arr = list(map(int, input().rstrip().split()))

    miniMaxSum(arr)
```

```python
# Strong Password

#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'minimumNumber' function below.
#
# The function is expected to return an INTEGER.
# The function accepts following parameters:
#  1. INTEGER n
#  2. STRING password
#

def minimumNumber(n, password):
    # Return the minimum number of characters to make the password strong
    a = b = p = d = 1
    for c in password:

        if c in "0123456789":
            a = 0
        if c in "abcdefghijklmnopqrstuvwxyz":
            b = 0
        if c in "ABCDEFGHIJKLMNOPQRSTUVWXYZ":
            p = 0
        if c in "!@#$%^&*()-+":
            d = 0
        
    counter = a+b+p+d
    
    while (n + counter) < 6:
        counter += 1
    
        
    return counter
if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input().strip())

    password = input()

    answer = minimumNumber(n, password)

    fptr.write(str(answer) + '\n')

    fptr.close()
```

```python
#Alternate Characters

#!/bin/python3

import math
import os
import random
import re
import sys
import itertools

#
# Complete the 'alternate' function below.
#
# The function is expected to return an INTEGER.
# The function accepts STRING s as parameter.
#

def alternate(s):
    # Write your code here
    uniq = []
    
    for i in s:
        if i not in uniq:
            uniq.append(i)
    
    if len(uniq) == 1:
        return 0
    
    else:
        combinations = list(itertools.combinations(uniq, len(uniq)-2))
        
        validTs = []
                
        for combo in combinations:
            
            t = s
            
            for char in combo:
                t = t.replace(char, "")
                
            flag = "up"    
            for i in range((len(t)-1)):
                if t[i] == t[i+1]:
                    flag = "down"
                    break
                
            if flag == "up":
                validTs.append(t)
        
        if validTs == []:
            return 0
        else:
            return max(map(len, validTs))
                
    

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    l = int(input().strip())

    s = input()

    result = alternate(s)

    fptr.write(str(result) + '\n')

    fptr.close()
```

```python
# Running Time

#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'runningTime' function below.
#
# The function is expected to return an INTEGER.
# The function accepts INTEGER_ARRAY arr as parameter.
#

def runningTime(arr):
    # Write your code here
    
    shifts = 0
    for i in range(1, len(arr)):
        j = i-1
        key = arr[i]
        while (j >= 0) and (arr[j] > key):
            shifts += 1
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key
    return shifts

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input().strip())

    arr = list(map(int, input().rstrip().split()))

    result = runningTime(arr)

    fptr.write(str(result) + '\n')

    fptr.close()
```

```python
# Quicksort 1 - Partition

#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'quickSort' function below.
#
# The function is expected to return an INTEGER_ARRAY.
# The function accepts INTEGER_ARRAY arr as parameter.
#

def quickSort(arr):
    # Write your code here
    
    p = arr[0]
    left = [arr[i] for i in range(len(arr)) if arr[i] < p]
    equal = [p]
    right = [arr[i] for i in range(len(arr)) if arr[i] > p]
    arr = left + equal + right
    
    return arr

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input().strip())

    arr = list(map(int, input().rstrip().split()))

    result = quickSort(arr)

    fptr.write(' '.join(map(str, result)))
    fptr.write('\n')

    fptr.close()
```

```python
#closest numbers

#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'closestNumbers' function below.
#
# The function is expected to return an INTEGER_ARRAY.
# The function accepts INTEGER_ARRAY arr as parameter.
#

def closestNumbers(arr):
    # Write your code here
    
    arr.sort()
    min_distance = abs(arr[0]-arr[1])
    pairs = []
    
    for i in range(len(arr)-1):
        distance = abs(arr[i] - arr[i+1])
        
        if distance < min_distance:
            pairs = [arr[i], arr[i+1]]
            min_distance = distance
            
        elif distance == min_distance:
            pairs.extend([arr[i],arr[i+1]])
            
    return pairs
        
        
    

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input().strip())

    arr = list(map(int, input().rstrip().split()))

    result = closestNumbers(arr)

    fptr.write(' '.join(map(str, result)))
    fptr.write('\n')

    fptr.close()
```

```python
# counting sort full

#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'countSort' function below.
#
# The function accepts 2D_STRING_ARRAY arr as parameter.
#

def countSort(arr):
    # Write your code here
    maxn = 0
    for i in arr:
        if int(i[0]) > maxn:
            maxn = int(i[0])
            
    freqArr = [ [] for i in range(maxn+1)]

    for i in range(len(arr)):
        
        if i < len(arr)//2:
            
            arr[i][1] = "-"
        
        freqArr[int(arr[i][0])].append((arr[i][1]))

    for i in freqArr:
        for j in i:
            print(j, end = " ")
            

if __name__ == '__main__':
    n = int(input().strip())

    arr = []

    for _ in range(n):
        arr.append(input().rstrip().split())

    countSort(arr)
```
